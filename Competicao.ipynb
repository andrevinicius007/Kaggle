```python
lab_Enc = LabelEncoder()
train['thal'] = lab_Enc.fit_transform(train['thal'])
train.head()
```

```python
test['thal'] = lab_Enc.fit_transform(test['thal'])
test.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
X = train.drop(['patient_id'],axis=1)
y = train_pred['heart_disease_present']
```

```python
modelo = DecisionTreeClassifier(random_state=1)
modelo.fit(X,y)
```

```python
params={
    'min_samples_split' : range(10,500,20),
    'max_depth': range(1,20,2)
}
grid = GridSearchCV(modelo,params)
grid.fit(X,y)
modelo = grid.best_estimator_
modelo.fit(X,y)
```

```python
recursos = train.columns.values

features_importance = zip(modelo.feature_importances_,recursos)
for importancia,recurso in sorted(features_importance, reverse=True):
    print("%s: %f%%" % (recurso, importancia*100))
```

```python
modelo.score(X,y)
```

```python
modelo_val = cross_val_score(modelo,X,y,cv=5)
print(modelo_val.mean())
```

```python
new_test = test.drop(['patient_id'],axis=1)
```

```python
predicao = modelo.predict_proba(new_test)
pred = np.array(predicao)
pred[:,0]
```

# Resolver problema do predict_proba q retorna uma lista de array

```python
submission = pd.DataFrame()
submission['patient_id'] = test['patient_id']
submission['heart_disease_present'] = pred[:,0]
submission.to_csv('submission.csv',index=False)
```
